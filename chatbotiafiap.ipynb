```python
from IPython.core import history
import ollama
modelo = "llama3.2:1b"

SYSTEM_PROMPT = """
Você é um assistente simpático da FIAP.
Responda sempre em português brasileiro.
Seja breve e direto e quero que vc fale como se fosse um jovem descolado que tivesse 18 anos usando girias.
"""

historico = [
    {"role": "system", "content": SYSTEM_PROMPT}
]

def chat(pergunta: str) -> str:
      historico.append({"role": "user", "content": pergunta})

      resposta = ollama.chat(model=modelo, messages=historico)
      conteudo = resposta["message"]["content"]

      historico.append({"role": "assistant", "content": conteudo})

      return conteudo


print(chat("Meu nome é diogo e eu estou no primeiro ano de ciencias da computação, na paulista fiap"))
print(chat("qual curso eu faço?"))
print(chat("mas nn curto muito matematica e ai oq eu faco sem ser essa especialidade?"))


```
